```julia
using Gadfly, DataFrames
```

```julia
T = 0:0.1:2pi
```

```julia
Y1 = sin.(T);
Y2 = cos.(T);
Y = vcat(Y1, Y2);
```

```julia
df = DataFrame(X=repeat(T, outer=[2]), Y=Y, index=repeat(["sin", "cos"], inner=[length(T)]));
```

```julia
plot(df, x=:X, y=:Y, color=:index, Geom.point, Geom.line, Guide.title("sin"))
```

```julia
Y_1 = randn(length(Y))./50 .+ Y;
```

```julia
df1 = DataFrame(X=repeat(T, outer=[2]), Y=Y_1, index=repeat(["sin", "cos"], inner=[length(T)]));
```

```julia
plot(df1, x=:X, y=:Y, color=:index, Geom.point, Geom.smooth)
```
